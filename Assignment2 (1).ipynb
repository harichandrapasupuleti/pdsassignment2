```python
# Check for missing values in each column
missing_values = df1.isnull().sum()
print("Missing Values:")
print(missing_values)
```

```python
# Handling missing values:


df1['Mileage'].fillna(df1['Mileage'].mean(), inplace=True)
df1['Engine'].fillna(df1['Engine'].mean(), inplace=True)
df1['Power'].fillna(df1['Power'].mean(), inplace=True)
df1['Seats'].fillna(df1['Seats'].mean(), inplace=True)

df1['New_Price'].fillna(df1['New_Price'].mean(), inplace=True)



df1['Location'].fillna(df1['Location'].mode()[0], inplace=True)
df1['Fuel_Type'].fillna(df1['Fuel_Type'].mode()[0], inplace=True)
df1['Transmission'].fillna(df1['Transmission'].mode()[0], inplace=True)

# Verify if there are no missing values left
print("Missing Values After Imputation:")
print(df1.isnull().sum())
```

```python


df1['Fuel_Type'].replace({'Diesel': 0, 'Petrol': 1, 'Electric': 2}, inplace=True)
df1['Transmission'].replace({'Manual': 0, 'Automatic': 1}, inplace=True)
```
